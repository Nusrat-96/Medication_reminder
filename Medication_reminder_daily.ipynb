```python
import datetime
import time
from dateutil import tz
```

```python
class Medication:
    def __init__(self):
        self.name = None
        self.dosage= None
        self.frequency = None
        self.time_of_first_dosage=None
        
    def medication_attributes(self, name, dosage, frequency, time_of_first_dosage):
        self.name = name
        self.dosage= dosage
        self.frequency = frequency
        self.time_of_first_dosage=time_of_first_dosage
        
        return self.__dict__
```

```python
class Patient:
    def __init__(self):
        self.patient_ID=None
        self.patient_name = None
        self.age = None
        self.address = None
        self.contact_info = None
        self.mdeications=[]
    
    def patient_information(self,patient_ID, patient_name, age, address, contact_info):
        self.patient_ID=patient_ID
        self.patient_name = patient_name
        self.age = age
        self.address = address
        self.contact_info = contact_info
        
    def add_medicaton(self, medication):
        self.mdeications.append(medication)
        
        
        
```

```python
class Medication_reminder(Patient, Medication):
    def __init__(self):
        super().__init__()
        self.patients =[]
        self.medications = []
        self.reminders=[]
        self.medication_reminder=[]
    
    
    def add_patients (self, patient_ID, patient_name, age, address, contact_info):
        super(Medication_reminder, self).patient_information( patient_ID, patient_name,
                                                             age, address, contact_info)
        self.patients.append([patient_ID, patient_name, age, address, contact_info])
        
        
    def add_medication (self, medication):
        self.medications.append(medication)
        super(Medication_reminder,self).add_medicaton(medication)
    
    

    
    def display_medication_schedule(self, patient):
        print("Medications for ", patient)
        print("------------------------")
        for medication in self.medications:
            for key, vlaue in medication.items():
                print(key,": ", vlaue)
            print("------------------------")
            
            
            
    
    #this function will add the dosage time for each medication depend on the dosage
    def add_dosage_time(self,medicine_name,dosage, initial_dosage_time):
        
        now = datetime.datetime.now()
        
        current_date = datetime.datetime.today().strftime("%Y-%m-%d")
        date_time_string= f"{current_date} {initial_dosage_time}"   #As manually we can't change the default year of strptime
        
        initial_time = datetime.datetime.strptime(date_time_string, "%Y-%m-%d %I:%M %p") #parses the string and returns a datetime object
        
        
        time_increase = datetime.timedelta(minutes=(24*60)/dosage)
        
        reminder_time = [initial_time+i*time_increase for i in range(0,dosage)]
           
        add_to_medication_reminder = {
            "medicine_name":medicine_name,
            "times_to_eat": reminder_time
        }
        self.medication_reminder.append(add_to_medication_reminder)
        print(self.medication_reminder)
        

      
    
    
    def check_reminder(self,medication_name, time_of_first_dosage):
        
        time_obj = datetime.datetime.strptime(time_of_first_dosage, "%H:%M")
        hour = time_obj.hour
        minute = time_obj.minute
        
        #now = datetime.now(tz.tzlocal())
        now = datetime.datetime.now()
        reminder ={
            "Medicine_Name": medication_name,
            "Time": datetime.time(hour=hour, minute=minute),
            "datetime": datetime.datetime(year=now.year, month=now.month, day=now.day, hour= hour, minute=minute)
        }
        
        self.reminders.append(reminder)
       
        #current_time= datetime.now(tz.tzlocal()).time()
        current_time = datetime.datetime.now()
        for reminder in self.reminders:
            reminder_time = reminder["datetime"]
            if current_time>=reminder_time:
                print("Reminder for : ", reminder["Medicine_Name"])
                reminder["datetime"] = reminder_time + datetime.timedelta(days=1)  #setting allert for the next day
            else:
                print("Not in time")
        
    
    


```

```python
medication1 = Medication()
medication1=medication1.medication_attributes("Asperin", "500mg", "two times of a day", "15:35")

medication2=Medication()
medication2=medication2.medication_attributes("Napa", "500mg", "two times of a day", "12:00")

reminder_system = Medication_reminder()
reminder_system.add_patients(43, "Nusrat", 37, "one-33", 232423)


reminder_system.add_medication(medication1)
reminder_system.add_medication(medication2)
reminder_system.display_medication_schedule (reminder_system.patient_name)

reminder_system.add_dosage_time("Asperin",3,"9:00 pm")

reminder_system.check_reminder(medication1["name"],"12:00")
reminder_system.check_reminder(medication2["name"],"12:00")


```
